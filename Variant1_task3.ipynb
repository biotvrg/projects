С помощью каких метрик можно оценить результаты последнего прошедшего события?

- Средняя дневная аудитория - среднее арифметическое дневной аудитории за определенный период. Используется, если нужно определить примерную дневную аудиторию продукта;
- MAU - количество уникальных пользователей, которые зашли в приложение в течение месяца;
- CR - отношение числа пользователей, которые выполнили какое-либо целевое действие к общему числу пользователей;
- Кол-во пользователей находящихся в игре в одно время;
- Conversion window (конверсионное окно) - время, которое пользователь тратит на переход на следующий уровень;
- ARPU – средний доход с привлеченного пользователя; за время прохождения акции
- LTV (lifetime value)  – пожизненная ценность клиента. Отражает то, сколько в среднем принесет пользователь денег в продукт до того, как покинет его; LTV до n-го дня (за время прохождения акции)
- Retention - метрика, которая показывает, как пользователи возвращаются в продукт (и часто свидетельствуют, нравится ли он им). за время прохождения акции в сравнении с обычным днем;
- Средняя продолжительность сессии (average session length, ASL) – среднее арифметическое длин всех сессий;
- Кол-во пройденных уровней.
